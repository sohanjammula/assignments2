```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

answer:

A scatter plot is a type of data visualization that displays the relationship between two variables by representing individual data points on a two-dimensional Cartesian coordinate system. Each data point is represented by a dot or marker, and the position of the dot on the plot corresponds to the values of the two variables.

In a scatter plot:

The horizontal axis (x-axis) represents one variable, often referred to as the independent variable.
The vertical axis (y-axis) represents another variable, often referred to as the dependent variable.
Each data point is plotted at the intersection of its corresponding values on the x-axis and y-axis.
Scatter plots are commonly used to:

Identify patterns or relationships between two continuous variables.
Visualize the distribution of data points.
Detect outliers or anomalies in the dataset.
Assess the strength and direction of the relationship between variables (positive, negative, or no correlation).
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
```

```python
Q3: Why is the subplot() function used?

answer:

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. Each subplot represents a separate 
plot area where you can visualize different datasets or different aspects of the same dataset.

Here are some reasons why the subplot() function is commonly used:

Organizing multiple plots: It allows you to arrange multiple plots in a grid layout, making it easier to compare different datasets or
visualize different aspects of the same data simultaneously.

Saving space: When you have several plots to display, arranging them in a grid layout using subplots can save space and make efficient use 
    of the figure area.

Comparing data: Subplots enable direct comparison between different datasets or variations of the same dataset, facilitating analysis and 
    interpretation.

Enhancing storytelling: Subplots can be used to present a series of related visualizations in a cohesive manner, allowing you to tell a story or
    convey complex information effectively.

Customization: Each subplot can be customized independently, including adjusting axes, titles, labels, colors, and other properties, providing 
    fine control over the appearance of individual plots.

Convenience: The subplot() function provides a simple interface for creating subplots, allowing you to specify the number of rows and columns in 
the grid and the position of each subplot within the grid.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='blue')
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='orange')
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='green')
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='red')
plt.title('Line Plot 4')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plots
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

answer:
A bar plot is a type of data visualization that represents categorical data with rectangular bars. The length or height of each bar corresponds to the
value of the data it represents. Bar plots are commonly used to compare the values of different categories or groups within a dataset.

Why is it used?

Comparison: Bar plots allow for easy comparison of quantities across different categories or groups.
Categorical data representation: They are suitable for visualizing categorical data, such as company names, product categories, or demographic groups.
Simple interpretation: Bar plots provide a straightforward visual representation of data, making it easy to interpret and understand the
underlying patterns or trends.
Presentation: They are commonly used in presentations, reports, and publications to visually communicate insights from the data.
Customization: Bar plots can be customized with different colors, labels, and styles to enhance their visual appeal and convey additional information.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Profit of Companies')
plt.show()
```

```python
# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

answer:

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a graphical summary of the distribution of a dataset. It is particularly useful for comparing the distribution of multiple datasets or identifying the presence of outliers.

Components of a Box Plot:

Median (Q2): The line inside the box represents the median value of the dataset.
Box (Q1 to Q3): The box represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). It contains the middle 50% of the data.
Whiskers: The lines (whiskers) extending from the box represent the range of the data. They typically extend to 1.5 times the IQR beyond the quartiles. Data points beyond the whiskers are considered outliers and plotted individually.
Outliers: Individual data points that fall beyond the whiskers of the box plot.
Why is it used?

Summary of distribution: Box plots provide a concise summary of the distribution of a dataset, including measures of central tendency (median) and spread (IQR).
Comparison: They are effective for comparing the distributions of multiple datasets or groups within a dataset.
Identification of outliers: Box plots visually highlight potential outliers in the data.
Robustness: Box plots are robust to outliers and skewed distributions, making them suitable for analyzing non-normal datasets.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a single list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Data')
plt.show()
```

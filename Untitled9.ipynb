```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

answer:
Scatter Plot:

Use: Visualize the relationship between two continuous variables.
Example: Exploring the correlation between height and weight.
Histogram:

Use: Display the distribution of a single continuous variable.
Example: Analyzing the distribution of ages in a population.
Bar Plot:

Use: Compare the quantities of different categories.
Example: Showing the sales performance of different products.
Box Plot:

Use: Summarize the distribution of a continuous variable and detect outliers.
Example: Analyzing the spread of salaries in a company.
Heatmap:

Use: Visualize the correlation or relationship between multiple variables in a matrix format.
Example: Displaying the correlation between different features in a dataset.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Show the plot
import matplotlib.pyplot as plt
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Adjust legend position
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Age Distribution Across Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Fare Distribution Across Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue parameter
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

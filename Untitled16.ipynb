```python
Q2: Differentiate between one-tailed and two-tailed tests.
```

```python
One-tailed and two-tailed tests are types of hypothesis tests that differ in the directionality of the hypothesis being tested and the critical region
for rejecting the null hypothesis.

One-Tailed Test:
In a one-tailed test, the null hypothesis (H0) and alternative hypothesis (Ha or H1) specify a direction for the effect or relationship being tested.
The critical region for rejecting the null hypothesis is located entirely in one tail of the sampling distribution (either the left tail or the right
tail), depending on the direction specified by the alternative hypothesis.
One-tailed tests are used when there is a specific directional hypothesis or when the researcher is only interested in detecting an effect in one 
    direction.
The decision to reject or fail to reject the null hypothesis is based solely on whether the observed data fall into the critical region in the
    specified direction.
Two-Tailed Test:
In a two-tailed test, the null hypothesis (H0) and alternative hypothesis (Ha or H1) do not specify a direction for the effect or relationship being 
    tested. Instead, they allow for the possibility of differences or effects in either direction.
The critical region for rejecting the null hypothesis is divided between both tails of the sampling distribution, with a specified level of 
significance (Œ±) allocated to each tail.
Two-tailed tests are used when there is no specific directional hypothesis or when the researcher wants to test for the possibility of differences
    or effects in both directions.
The decision to reject or fail to reject the null hypothesis is based on whether the observed data fall into either of the two critical regions.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
```

```python

A confidence interval is a range of values derived from sample data that is used to estimate the true population parameter with a certain level 
of confidence. It provides a range of plausible values for the population parameter, rather than a single point estimate. The confidence level 
associated with a confidence interval represents the probability that the interval contains the true population parameter.

Here's how to calculate a confidence interval:

Choose the Confidence Level: Determine the desired level of confidence for the interval, typically expressed as a percentage (e.g., 95%, 99%).
Select the Sample and Collect Data: Obtain a random sample from the population of interest and collect the necessary data.
Calculate the Sample Statistic: Compute the sample statistic (e.g., sample mean, sample proportion) that estimates the population parameter.
Determine the Margin of Error: Calculate the margin of error, which is based on the variability of the sample statistic and the desired level of 
confidence.
Construct the Confidence Interval: Use the sample statistic and the margin of error to construct the confidence interval.
```

```python

The margin of error in a confidence interval is a measure of the precision or uncertainty associated with the estimate of the population parameter. 
It represents the maximum likely difference between the sample statistic (e.g., sample mean) and the true population parameter within the confidence
interval.

In the context of estimating the population mean (
ùúá
Œº) with a confidence interval, the margin of error is calculated using the standard error of the sample mean. For a confidence interval for the 
population mean when the population standard deviation (
ùúé
œÉ) is known, the margin of error (
ùê∏
E) is calculated as:

E=z√ó(œÉ/sqrt(n))

Where:

ùëß
z is the critical value from the standard normal distribution corresponding to the desired confidence level.
ùúé
œÉ is the population standard deviation.
ùëõ
n is the sample size.
The margin of error decreases with increasing sample size (
ùëõ
n). As the sample size increases, the standard error decreases, resulting in a smaller margin of error. This is because larger sample sizes provide
more information about the population, leading to a more precise estimate of the population parameter.
```

```python
import requests
from bs4 import BeautifulSoup

# Function to extract video URLs
def extract_video_urls(channel_url, num_videos):
    # Fetch the webpage content
    response = requests.get(channel_url)
    if response.status_code != 200:
        print("Failed to fetch the webpage")
        return []

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all links to videos
    video_links = soup.find_all('a', class_='yt-simple-endpoint style-scope ytd-grid-video-renderer')

    # Extract URLs of the first num_videos videos
    video_urls = []
    for link in video_links[:num_videos]:
        video_url = 'https://www.youtube.com' + link['href']
        video_urls.append(video_url)

    return video_urls

# URL of the YouTube channel
channel_url = "https://www.youtube.com/@PW-Foundation/videos"

# Number of videos to extract
num_videos = 5

# Extract video URLs
video_urls = extract_video_urls(channel_url, num_videos)

# Print the extracted video URLs
for i, url in enumerate(video_urls, 1):
    print(f"Video {i}: {url}")
```

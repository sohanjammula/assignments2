```python
Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.
```

```python
import math

def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
        x (float): The point at which to calculate the PDF.
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.

    Returns:
        float: The PDF value at the given point x.
    """
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)
    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution
point = 1  # Point at which to calculate the PDF
pdf_value = normal_pdf(point, mean, std_dev)
print("PDF at x =", point, ":", pdf_value)
```

```python
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.
```

```python

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, 
where each trial has only two possible outcomes: success or failure. The properties of the binomial distribution are as follows:

Fixed Number of Trials: The binomial distribution describes the number of successes in a fixed number of independent trials, denoted by 
�
n.

Independent Trials: Each trial is independent of the others. The outcome of one trial does not affect the outcome of another trial.

Two Possible Outcomes: Each trial has only two possible outcomes: success or failure.

Constant Probability of Success: The probability of success, denoted by 
�
p, remains constant from trial to trial.

Discrete Distribution: The binomial distribution is discrete, meaning it deals with countable outcomes (whole numbers).

Parameters: The parameters of the binomial distribution are 
�
n, the number of trials, and 
�
p, the probability of success in each trial.

Examples of events where the binomial distribution can be applied include:

Coin Flipping: Consider flipping a fair coin 
�
n times and counting the number of heads. Each flip is an independent trial with two possible outcomes (heads or tails), and the probability of getting 
a head is 
�
=
0.5
p=0.5.

Manufacturing Quality Control: Suppose a manufacturing process produces electronic components, and each component has a probability 
�
p of being defective. If we randomly sample 
�
n components from the production line and count the number of defective components, the number of defectives follows a binomial distribution.
This applies as long as the components are produced independently and the probability of a defective component remains constant.

These examples illustrate situations where the binomial distribution is applicable due to the fixed number of trials, independence of trials, and two 
possible outcomes with a constant probability of success.
```

```python
Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.
```

```python
import math

def poisson_cdf(k, lambd):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
        k (int): The point at which to calculate the CDF.
        lambd (float): The mean of the Poisson distribution.

    Returns:
        float: The CDF value at the given point k.
    """
    cdf = 0
    for i in range(k + 1):
        cdf += (lambd ** i) * math.exp(-lambd) / math.factorial(i)
    return cdf

# Example usage:
mean = 2  # Mean of the Poisson distribution
point = 3  # Point at which to calculate the CDF
cdf_value = poisson_cdf(point, mean)
print("CDF at k =", point, ":", cdf_value)
```

```python
Q7. How Binomial distribution different from Poisson distribution?
```

```python
Binomial and Poisson distributions are both used to model the number of occurrences of events, but they differ in their underlying assumptions and the types of
scenarios they are appropriate for.

Assumptions:

Binomial Distribution: The binomial distribution is used when there are a fixed number of independent trials, each with two possible outcomes 
                              (success or failure), and the probability of success remains constant for each trial. The outcomes of trials are discrete
                                                                                                   and independent.
Poisson Distribution: The Poisson distribution is used when events occur at a constant average rate over a fixed interval of time or space. It assumes 
that the events are independent and occur randomly.
Number of Possible Outcomes:

Binomial Distribution: The binomial distribution deals with a fixed number of trials with two possible outcomes (success or failure) in each trial.
Poisson Distribution: The Poisson distribution deals with events that can occur any number of times within a fixed interval (e.g., number of phone calls 
                    received in an hour, number of arrivals at a restaurant).
Parameterization:

Binomial Distribution: The parameters of the binomial distribution are the number of trials (
�
n) and the probability of success in each trial (
�
p).
Poisson Distribution: The parameter of the Poisson distribution is the average rate of event occurrences (
�
λ) over the fixed interval.
Limiting Behavior:

As the number of trials (
�
n) in a binomial distribution increases while keeping the product 
�
�
np constant, the binomial distribution approaches the shape of a Poisson distribution. This is known as the Poisson limit of the binomial distribution.
Applicability:

Binomial Distribution: Typically applied in scenarios with a fixed number of trials and binary outcomes, such as coin flips, pass/fail experiments, or the 
number of successes in a series of independent Bernoulli trials.
Poisson Distribution:
Typically applied in scenarios where events occur randomly and independently over a fixed interval of time or space, such as the number of arrivals at a 
service center, the number of phone calls received in a call center during a specific time period, or the number of accidents at an intersection in a given
time frame.

In summary, while both the binomial and Poisson distributions deal with the count of events, they differ in their assumptions about the nature of the events 
and the circumstances under which they occur. The choice between the two distributions depends on the specific characteristics of the data and the underlying
    process being modeled.
```

```python
Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.
```

```python
import numpy as np

# Parameters
sample_size = 1000
mean = 5

# Generate random sample from Poisson distribution
random_sample = np.random.poisson(mean, sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

```python
Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
```

```python
In a normal distribution, also known as a Gaussian distribution, the probability density function (PDF) is symmetric around the mean. 
This means that the probability of observing data decreases as you move away from the mean in either direction. Therefore, the least frequent data appears 
at the tails of the distribution, farthest away from the mean position.

In a standard normal distribution with mean 

μ=0 and standard deviation 

σ=1, the least frequent data appears at the extreme ends of the distribution, or more precisely, in the tails beyond a certain number of standard deviations from the mean.

In a normal distribution:

Approximately 68% of the data falls within one standard deviation (σ) of the mean (μ).
Approximately 95% of the data falls within two standard deviations of the mean.
Approximately 99.7% of the data falls within three standard deviations of the mean.
Therefore, the least frequent data appears beyond three standard deviations from the mean in either direction. These extreme values represent the 
outliers or rare occurrences in the distribution.
```
